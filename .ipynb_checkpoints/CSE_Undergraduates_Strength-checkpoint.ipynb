```python
# Get the list of numeric columns (excluding 'Strength' and 'Overall_Score')
numeric_columns = df.select_dtypes(include=['number']).columns

# Create histograms for each numeric column
for column in numeric_columns:
    plt.figure(figsize=(8, 4))
    df[column].plot(kind='hist', bins=20, edgecolor='k', title=f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Select a category column and target variable
# Get a list of category columns (replace these with your actual category column names)
category_columns = [
    "Structured Programming",
    "Structured programming (Sessional)",
    "OOP",
    "OOP (Sessional)",
    "Discrete Math",
    "Calculus",
    "English Skill Development",
    "Data Structure",
    "Data Structure (Sessional)",
    "Algorithm",
    "Algorithm (Sessional)",
    "Applied Statistics",
    "CP Lab",
    "DBMS",
    "DBMS (Sessional)",
    "Theory of Computing",
    "Operating Systems",
    "Operating Systems (Sessional)",
    "Computer Architecture",
    "Artificial Intelligence",
    "Machine Learning",
    "Compiler Design",
    "Engineering Management",
    "Thesis",
    "Total Problems Solved"
]

# Specify the target variable
target_variable = 'Strength'

# Create histograms for each category column versus target variable
for category_column in category_columns:
    plt.figure(figsize=(10, 6))
    for category_value in df[category_column].unique():
        subset = df[df[category_column] == category_value]
        sns.histplot(subset[target_variable], label=f'{category_column} = {category_value}', kde=True)
    
    plt.title(f'Histograms of {target_variable} by {category_column}')
    plt.xlabel(target_variable)
    plt.legend()
    plt.show()

# Create a heatmap to visualize correlations
correlation_matrix = df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, cbar=True, square=True, fmt=".1f", annot=True, annot_kws={'size':10}, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Assuming the last column 'Strength' is the target variable
X_ml = df.iloc[:, :-1]  # Excluding the last two columns (Overall_Score and Strength)
y_ml = df['Strength'].astype('int')  # Ensure it's an integer type
```

```python
print(f"Shape of each variables- Category Variables: {X_ml.shape} and Target Variable: {y_ml.shape}")
```

```python
print(f"Top values of each variables- Category Variables: {X_ml.head()} and Target Variable: {y_ml.head()}")
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_ml, y_ml, test_size=0.2, random_state=42)
```

```python
# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize machine learning models
log_reg = LogisticRegression(max_iter=1000)
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
```

```python
# Train and evaluate each model
models = {'Logistic Regression': log_reg, 'Random Forest': rf, 'Gradient Boosting': gb}
results = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    results[name] = classification_report(y_test, y_pred, output_dict=True)
    
    # Calculate and display the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
# Display the classification report for the Logistic Regression model
log_reg_report = classification_report(y_test, log_reg.predict(X_test_scaled))
print(log_reg_report)
```

```python
# Display the classification report for the Random Forest Regression model
rf_report = classification_report(y_test, rf.predict(X_test_scaled))
print(rf_report)
```

```python
# Display the classification report for the Gradient Boosting model
gb_report = classification_report(y_test, gb.predict(X_test_scaled))
print(gb_report)
```

```python
# Therefore, the Logistic Regression Model is most suitable
```

```python
# Dictionary to store model filenames
model_filenames = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    results[name] = classification_report(y_test, y_pred, output_dict=True)
    
    # Save each model to a separate file
    model_filename = f'{name.lower().replace(" ", "_")}_model.pkl'
    joblib.dump(model, model_filename)
    model_filenames[name] = model_filename
```
